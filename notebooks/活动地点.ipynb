```python
from geopy.geocoders import GoogleV3

# Initialize geolocator with Google API key
geolocator = GoogleV3(api_key='YOUR_GOOGLE_API_KEY')

def get_coordinates(location):
    location = geolocator.geocode(location + ', New York, NY')
    if location:
        return (location.latitude, location.longitude)
    else:
        return (None, None)

# Apply the function to your DataFrame as before
df[['Latitude', 'Longitude']] = df['Location Details'].progress_apply(lambda x: pd.Series(get_coordinates(x)))

print(df.head())
```
